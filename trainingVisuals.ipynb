```python
import torch
import os
import matplotlib.pyplot as plt
```

```python
def plot_training_metrics(model_root_path, models, colors, title="Training Metrics"):    
    # Plot 2 figures, one for loss and one for accuracy
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(title)

    for i, model in enumerate(models):
        path = os.path.join(model_root_path, model+".pth")
        checkpoint = torch.load(path)
        log = checkpoint['log']

        logTrain = [entry['train'] for entry in log]
        logTrainLoss = [entry['loss'] for entry in logTrain]
        logTrainAcc = [entry['acc'] for entry in logTrain]
        logVal = [entry['val'] for entry in log]
        logValLoss = [entry['loss'] for entry in logVal]
        logValAcc = [entry['acc'] for entry in logVal]

        ax[0].plot(logTrainLoss, label=f"{model} Train", color=colors[i])
        ax[0].plot(logValLoss, label=f"{model} Validation", linestyle="--", color=colors[i])

        ax[1].plot(logTrainAcc, label=f"{model} Train", color=colors[i])
        ax[1].plot(logValAcc, label=f"{model} Validation", linestyle="--", color=colors[i])

        # Print final values
        print(f"Final values for {model}:")
        print("Train Loss:", logTrainLoss[-1])
        print("Train Accuracy:", logTrainAcc[-1])
        print("Validation Loss:", logValLoss[-1])
        print("Validation Accuracy:", logValAcc[-1])

    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Loss")
    ax[0].set_title("Loss")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Accuracy")
    ax[1].set_title("Accuracy")

    # put one legend on the right of the rightmost axis
    ax[-1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
# Get current working directory
model_root_path = os.path.join(os.getcwd(), "models")
models = ["ResNetA", "ResNetB"]
colors = ["red", "blue"]

plot_training_metrics(model_root_path, models, colors, title="Training Metrics for ResNet Partitions")
```

```python
models = ["VimA", "VimB"]
colors = ["red", "blue"]

plot_training_metrics(model_root_path, models, colors, title="Training Metrics for ViM Partitions")
```

```python
models = ["VitA", "VitB"]
colors = ["red", "blue"]

plot_training_metrics(model_root_path, models, colors, title="Training Metrics for ViT Partitions")
```

```python
models = ["VimA", "VimB", "ResNetA", "ResNetB", "VitA", "VitB"]
colors = ["red", "orange", "blue", "green", "black", "brown"]
plot_training_metrics(model_root_path, models, colors, title="Training Metrics for All Model Partitions")
```
